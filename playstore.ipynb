```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset
data = pd.read_csv("/content/Google-Playstore.csv")
```

```python
#Checking for missing values
data.isnull().sum()
```

```python
#checking for duplicates
data.duplicated().sum()
```

```python
# Drop rows where 'App Name' is missing (critical field)
data.dropna(subset=['App Name'], inplace=True)
```

```python
# Fill missing 'Currency' with the most common currency ('USD')
data['Currency'].fillna(data['Currency'].mode()[0], inplace=True)
```

```python
# Fill missing 'Minimum Android' with the most common version
data['Minimum Android'].fillna(data['Minimum Android'].mode()[0], inplace=True)

# Fill missing 'Developer Id', 'Developer Website', and 'Developer Email'
data['Developer Id'].fillna("Unknown", inplace=True)
data['Developer Website'].fillna("Not Available", inplace=True)
data['Developer Email'].fillna("Unknown", inplace=True)
```

```python
# Convert 'Last Updated' to datetime
data['Last Updated'] = pd.to_datetime(data['Last Updated'], errors='coerce')

# Fill missing 'Released' with Unknown
data['Released'].fillna("Unknown", inplace=True)

print(data['Last Updated'].dtype)
```

```python
# Converting 'Released' column to datetime format
data['Released'] = pd.to_datetime(data['Released'], errors='coerce')
print(data['Released'].dtype)
print(data['Released'].head())
```

```python
# Fill missing 'Privacy Policy' with "Not Available"
data['Privacy Policy'].fillna("Not Available", inplace=True)

# Convert 'Scraped Time' to datetime
data['Scraped Time'] = pd.to_datetime(data['Scraped Time'], errors='coerce')
```

```python
# Fill missing Ratings with 0 where Rating Count is 0, otherwise use median by category
data['Rating'] = data.groupby('Category')['Rating'].transform(lambda x: x.fillna(x.median()))
data['Rating'].fillna(0, inplace=True)  # If still NaN, replace with 0

# If Rating Count is missing, set to 0
data['Rating Count'].fillna(0, inplace=True)
```

```python
print(data[['Rating', 'Rating Count']].isnull().sum())
```

```python
# Function to clean and convert 'Size' column
def convert_size(size):
    if isinstance(size, str):  # Ensure size is a string
        size = size.replace(",", "")  # Remove commas

        if "M" in size:
            return float(size.replace("M", ""))  # Convert MB
        elif "K" in size:
            return float(size.replace("K", "")) / 1024  # Convert KB to MB
    return np.nan  # Return NaN for missing or unknown values

# Apply the function to the 'Size' column
data['Size'] = data['Size'].apply(convert_size)

# Fill NaN values with the median size and convert to float32
data['Size'] = data['Size'].fillna(data['Size'].median()).astype('float32')

# Verify the results
data['Size'].unique()[:10]  # Check some unique values
data.info()  # Confirm the datatype of 'Size'
```

```python
# Function to clean the "Installs" column
def clean_installs(value):
    if isinstance(value, str):
        return int(value.replace('+', '').replace(',', ''))  # Remove '+' and ',' then convert to int
    return value  # Return value as-is if already a number

# Clean the "Installs" column
data['Installs'] = data['Installs'].apply(clean_installs)

# Convert to numeric type
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')

# Fill missing Installs with Minimum Installs where available
data['Installs'].fillna(data['Minimum Installs'], inplace=True)

# Fill remaining missing values with median installs per category
data['Installs'] = data.groupby('Category')['Installs'].transform(lambda x: x.fillna(x.median()))

# Final check and fill remaining with 0
data['Installs'].fillna(0, inplace=True)

# Ensure Installs is integer type
data['Installs'] = data['Installs'].astype('Int64')

# Check if null values remain
print(data['Installs'].isnull().sum())
```

```python
# Drop unnecessary columns
columns_to_drop = [
    "Scraped Time", "Developer Email", "Developer Id",
    "Minimum Installs", "Maximum Installs", "Free"
]

data_cleaned = data.drop(columns=columns_to_drop)
```

```python
# Drop "developer website" column
data.drop(columns=['Developer Website'], inplace=True, errors='ignore')
# Drop "Developer Email" column
data.drop(columns=['Developer Email'], inplace=True, errors='ignore')
# Drop "Privacy Policy" column
data.drop(columns=['Privacy Policy'], inplace=True, errors='ignore')
```
